```python
import os
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
import re
```

```python
IMAGE_PATH = '/prueba.jpg'
reader = easyocr.Reader(['es'], gpu=False)
result = reader.readtext(IMAGE_PATH)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
spacer = 100
for detection in result:
    op_left = tuple(result[0][0][0])
    text = detection[1]

    # Calcular las coordenadas de bottom_right para hacer el rectángulo más largo
    text_width = len(text) * 30  # Ajusta el valor 10 según tus necesidades
    bottom_right = (top_left[0] + text_width, top_left[1] + 70 )  # Ajusta el valor 20 según tus necesidades

    img = cv2.imread(IMAGE_PATH)
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15
plt.figure(figsize=(20,10))
plt.imshow(img)
plt.show()



```
